```python
import copy

#对data数据预处理
from scipy.io import arff
DAG= pd.read_csv("data/real_world_datasets/Hierarchy/LDStumbleupon.csv",header=None,error_bad_lines=False)
file_name='data/real_world_datasets/Datasets/LDStumbleupon.arff'
data,meta=arff.loadarff(file_name)
df=pd.DataFrame(data)

#从bytes转为int
for c in df.columns:
    if(c == "GI_value"):
        df[c] = df[c].apply(lambda x:bytes.decode(x)).replace({"true":1,"false":0})
    else:
        df[c] = df[c].apply(lambda x:(int.from_bytes(x, byteorder='big', signed=False))-48)
        
#处理多余的\
c = df.columns.tolist()
for i in range(len(c)):
    s = c[i]
    if('\\' in c[i]):
        c[i] = c[i].replace('\\','')
df.columns = c 
df.to_csv("data/real_world_datasets/Datasets/DATA_Stumbleupon.csv",index=None)
```

```python
#DAG中dag[i][j] = 1 表示第i个节点指向第j个节点，即i是j的父亲节点
df= pd.read_csv("data/real_world_datasets/Hierarchy/LDStumbleupon.csv",header=None,error_bad_lines=False)
data= pd.read_csv("data/real_world_datasets/Datasets/DATA_Stumbleupon.csv",header=0)
nodesName = data.columns[0:-1]

#构造DAG

DAG = pd.DataFrame(0,columns=nodesName,index=nodesName)
num_row,num_column = df.shape[0],df.shape[1]
for i in range(num_row):
    for j in range(1,num_column):
        if(pd.isnull(df.loc[i][j])):
            continue
        try:
            x = df.loc[i][j].strip()
            y = df.loc[i][0].strip()
            if(DAG.loc[x,y] == 0):
                DAG.loc[x,y] = 1
        except:
            pass
DAG.to_csv("data/real_world_datasets/Hierarchy/DAG_Stumbleupon.csv")
```
